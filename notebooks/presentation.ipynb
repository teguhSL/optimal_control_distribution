```python
import matplotlib.pyplot as plt
import numpy as np
```

#### 2D Plot

```python
N = 1000
H1 = 1
H2 = 2
H3 = 5
x = np.linspace(-10, 10, N)
y1 = 0.5*x*H1*x
y2 = 0.5*x*H2*x
y3 = 0.5*x*H3*x
plt.plot(x, y1, 'r', label='H = '+str(H1))
plt.plot(x, y2, 'g', label='H = '+str(H2))
plt.plot(x, y3, 'b', label='H = '+str(H3))
plt.xlabel('x', fontsize=15)
h = plt.ylabel('f(x)', fontsize=15)
h.set_rotation(0)
plt.ylim([0,100])
plt.legend()
plt.savefig('figures/quadratic.pdf', dpi=500, bbox_inches='tight')
```

#### Gaussian Counterpart

```python
Sigma1 = 1./H1
Sigma2 = 1./H2
Sigma3 = 1./H3
y1 = (1./np.sqrt(2*np.pi*Sigma1))*np.exp(-0.5*x*H1*x)
y2 = (1./np.sqrt(2*np.pi*Sigma2))*np.exp(-0.5*x*H2*x)
y3 = (1./np.sqrt(2*np.pi*Sigma3))*np.exp(-0.5*x*H3*x)
plt.plot(x, y1, 'r', label='Var = '+str(Sigma1))
plt.xlabel('x', fontsize=15)
h = plt.ylabel('p(x)', fontsize=15)
h.set_rotation(0)
plt.plot(x, y2, 'g', label='Var = '+str(Sigma2))
plt.plot(x, y3, 'b', label='Var = '+str(Sigma3))
plt.legend()
plt.savefig('figures/gaussian.pdf', dpi=500, bbox_inches='tight')
# plt.ylim([0,100])
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-20, 20, 0.25)
Y = np.arange(-20, 20, 0.25)
X, Y = np.meshgrid(X, Y)
Z = 2*X**2 + Y**2

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
# ax.set_zlim(-1.01, 1.01)
# ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
# ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
